```python
import pandas as pd
```

```python
import sklearn
```

```python
data = pd.read_csv('P2P_Lending.csv')
```

```python
data
```

```python
data=data.dropna()
```

```python
df = data.copy()
```

```python
target = 'int_rate'
```

```python
encode = ['TermInMonths','emp_length','home_ownership','loan_status','purpose']
```

```python
from sklearn import preprocessing 
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder = LabelEncoder()
label = preprocessing.LabelEncoder() 
```

```python
df['TermInMonths']= label.fit_transform(df['TermInMonths'])
df['emp_length']= label.fit_transform(df['emp_length']) 
df['home_ownership']= label.fit_transform(df['home_ownership']) 
df['loan_status']= label.fit_transform(df['loan_status']) 
df['purpose']= label.fit_transform(df['purpose']) 
```

for col in encode:
    dummy = pd.get_dummies(df[col], prefix=col)
    df = pd.concat([df,dummy], axis=1)
    del df[col]

```python
X = df.drop('int_rate', axis=1)
Y = df['int_rate']
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
model.fit(X, Y)
```

```python
for col in X.columns:
    print(col)
```

```python
import pickle
pickle.dump(model, open('intreglab.pkl', 'wb'))
```
